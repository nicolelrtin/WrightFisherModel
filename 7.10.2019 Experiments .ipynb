```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
%matplotlib inline
```

```python
clones = np.array([[90, 288, 49, 47, 35, 39, 27, 36, 48, 40, 41, 55], 
                    [146, 374, 60, 45, 43, 47, 51, 42, 46, 40, 66, 76], 
                    [333, 422, 68, 62, 52, 40, 41, 53, 46, 69, 95, 134],
                    [239, 367, 124, 76, 84, 86, 90, 87, 106, 119, 123, 117]])
wild = np.array([[259, 10, 259, 395, 283, 295, 337, 265, 369, 276, 344, 266],
                [252, 22, 311, 284, 292, 294, 302, 292, 282, 320, 338, 315],
                [458, 14, 338, 309, 311, 221, 250, 249, 267, 245, 263, 387],
                [269, 23, 243, 249, 440, 452, 353, 338, 380, 386, 389, 372]])

# f = array/total
log = np.log(10**(-10)+f/(1-f))


lg = lambda f: np.log(10**(-10)+f / (1-f))
x = range(4)*np.log(100)/np.log(2)

def polyfits(clones, total):
    total = clones + wild
    f = clones / (clones+wild)
    log = lg(f)
        
    tn1 = [log[0][:3]] + [log[1][:3]] + [log[2][:3]] + [log[3][:3]]
    tn2 = [log[0][3:6]] + [log[1][3:6]] + [log[2][3:6]] + [log[3][3:6]]
    tn3 = [log[0][6:9]] + [log[1][6:9]] + [log[2][6:9]] + [log[3][6:9]]
    tn4 = [log[0][9:12]] + [log[1][9:12]] + [log[2][9:12]] + [log[3][9:12]]
    
    fig, ax = plt.subplots(1, 4, tight_layout=True, figsize=(18, 4.5))
    for i, tn in enumerate([tn1, tn2, tn3, tn4]):
        fitness=np.polyfit(x, tn, 1)[0]
        ax[i].plot(x, tn, '-b', linewidth=1)
        ax[i].set_ylim([-2.5,4])
        ax[i].set_title('Clone {}, median fitness: {}'.format(i+1,np.around(np.median(fitness),3)))
        ax[i].set_xlabel('Time, generations')
        if i == 0:
            ax[i].set_ylabel(r'logit$(f_{mut})$')
    plt.show()

polyfits(clones, total)
```
